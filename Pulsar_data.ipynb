```r
library(tidyverse)
library(repr)
library(readxl)
library(readr)
```

```r
pulsar_data <- read_csv("HTRU_2.csv", col_names =  c("mean_int",
                        "sd_int",
                        "excess_int",
                        "skewness_int",
                        "mean_DMSNR",
                        "sd_DMSNR",
                        "excess_DMSNR",
                        "skewness_DMSNR",
                        "Class")
                       )
pulsar_data <- pulsar_data |>
                mutate( "Class" = as_factor(Class))
```

```r
head(pulsar_data)
```

```r
int_data <- pulsar_data |>
        select(mean_int:skewness_int, Class)
DMSNR_data <- pulsar_data |>
        select(mean_DMSNR:skewness_DMSNR, Class)

head(int_data)
head(DMSNR_data)
```

```r
positive_int <- int_data |>
                filter(Class =="1")
(positive_int)
```
